```python
import pandas as pd
```

```python
df_laptop = pd.read_csv('laptop_price.csv')
df_laptop.head()
```

# Filter dataframe based on a condition

```python
# find which rows have "Apple" in the "Company" column
df_laptop['Company'] == 'Apple'
```

```python
# filter dataframe based on condition
df_laptop[df_laptop['Company'] == 'Apple']
```

```python
# verify which companies are in the "Company" column with value_counts()
df_laptop[df_laptop['Company'] == 'Apple'].value_counts('Company')
```

# Filter dataframe based on multiple conditions

```python
# find Apple laptops that cost more than 2000 euros
(df_laptop['Company']=='Apple') & (df_laptop['Price_euros']>2000)
```

# & (and)

```python
df_laptop[(df_laptop['Company']=='Apple') & (df_laptop['Price_euros']>2000)]
```

```python
df_laptop[(df_laptop['Company']=='Apple') & (df_laptop['Price_euros']>2000)].value_counts('Company')
```

# | (or)

```python
# find laptops Apple or Dell laptops
df_laptop['Company']=='Apple'
df_laptop['Company']=='Dell'

(df_laptop['Company']=='Apple') | (df_laptop['Company']=='Dell')
```

```python
df_laptop[(df_laptop['Company']=='Apple') | (df_laptop['Company']=='Dell')]
```

```python
df_laptop[(df_laptop['Company']=='Apple') | (df_laptop['Company']=='Dell')].value_counts('Company')
```
