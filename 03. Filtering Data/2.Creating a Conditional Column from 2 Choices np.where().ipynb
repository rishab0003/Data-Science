```python
import pandas as pd
df_lap = pd.read_csv('laptop_price.csv')
df_lap.head()
```

# Creating a Conditional Column from 2 Choices: np.where()


```python
import numpy as np
```

```python
np.where(df_lap['Price_euros'] > 2000 , 'Expensive', 'Cheap')
```

```python
# set it to a new column
df_lap["Price_tier"] = np.where(df_lap['Price_euros'] > 2000 , 'Expensive', 'Cheap')
```

```python
df_lap.head()
```

```python
df_lap.value_counts("Price_tier")
```

# Creating a Conditional Column from More Than 2 Choices: np.select()

```python
# create an array based on multiple price tiers (+2 choices)

# create "conditions" and "values" variables
conditions = [
    (df_lap['Price_euros']>3000),
    (df_lap['Price_euros']>2000) & (df_lap['Price_euros']<=3000),
    (df_lap['Price_euros']>800) & (df_lap['Price_euros']<=2000),
    (df_lap['Price_euros']<=800),
]
choices = ['Too Expensive', 'Expensive', 'Affordable', 'Cheap']
```

```python
# set it to a new column
df_lap['Price_tier'] = np.select(conditions, choices, default='Unknown')
```

```python
# count values inside "price_tier" column
df_lap.value_counts('Price_tier')
```
