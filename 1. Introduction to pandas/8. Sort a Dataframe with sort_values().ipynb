```python
import pandas as pd
df_exam = pd.read_csv('StudentsPerformance.csv')
df_exam.head()
```

# Sort a dataframe

```python
df_exam.sort_values('math score')
```

```python
# sort descending by multiple columns
df_exam.sort_values('math score', ascending=False)
```

```python
# sort descending by multiple columns and update dataframe
df_exam.sort_values(['math score', 'reading score'], ascending=False, inplace=True)
```

```python
df_exam.head()
```

```python
# sort descending with a key function
df_exam.sort_values(['race/ethnicity'], ascending=True, key=lambda a:a.str.lower())
```

```python
df_exam.sort_values('race/ethnicity',ascending=False)
```
